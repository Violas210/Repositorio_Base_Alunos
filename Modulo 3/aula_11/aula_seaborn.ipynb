## O Que é o Seaborn ?

É uma biblioteca para criar visualizações estatisticas em python com pousas linhas de codigo
obs. Necessita do matplotlib. 

```python
import matplotlib.pyplot as plt
import seaborn as sns

# vamos usar um dataset de exemplo da própria biblioteca
#Exemplos : gorjetas, pinguins, titanic, diamantes, iris

gorjetas = sns.load_dataset('tips')

#display(gorjetas.head())

#print(gorjetas.shape)

for i in gorjetas:
    print(i)
```

```python
# Grafico de dispersão

sns.scatterplot(
    data=gorjetas,
    x='total_bill',
    y='tip',
    hue='sex',

)
plt.title('Relação entre total da conta e as gorjetas')
plt.show()
```

```python
# GRáfico de barras

sns.barplot(data=gorjetas,x='day',y='tip', hue='tip')
plt.show()
```

```python
# gráfico de hístograma

sns.histplot(
    data=gorjetas,
    x='tip',
    kde=True
)
plt.title('distribuição de gorjetas')
plt.show()
```

```python
pinguins = sns.load_dataset('penguins')

display(pinguins.head())

#for i in pinguins:
#    print(i)
```

```python
sns.set_theme(style='whitegrid')
sns.scatterplot(data=pinguins,x='body_mass_g',y='bill_depth_mm' ,hue='species')
plt.title('dispersão entre peso, comprimento do bico por esepcie')
#importante salvar antes do plt.show
plt.savefig(
    'grafico_dispersão.png',
    dpi = 300
)

plt.show()
```

```python
# Crie um Gráfico de coluna entre peso e especie

sns.barplot(
    data=pinguins,
    x='species',
    y='body_mass_g',
    hue='sex',

)
plt.title('distribuição de peso')
plt.show()
```

```python
sns.histplot(
    data=pinguins,
    x='body_mass_g',
    hue='species',
    kde=True
)
plt.savefig(
    'grafico_distribuição de peso por especies.png',
    dpi = 300
)
plt.title('distribuição de peso por especie')
plt.show()
```

```python
sns.boxplot(
    data=pinguins,
    x='species',
    y='body_mass_g',
    hue='island',
)
```
