# Testar visualizações com dois datasets do seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
titanic = sns.load_dataset('titanic')

display(titanic.head())

for i in titanic:
    print(i)
```

```python
sns.violinplot(
    data=titanic,
    x='class',
    y='survived',
    hue='sex',


)
plt.title('')
plt.show()
```

```python
sns.histplot(
    data=titanic,
    x='age',
    kde=True,
)
plt.title('idade que mais faleceram')
plt.show()
```
