```python
import numpy as np
import time

# ---

# 2. Gerar arrays grandes para o teste
# Vamos criar dois arrays NumPy com 50 milhões de elementos.
# Este tamanho é ideal para que a diferença de performance seja perceptível.
ARRAY_SIZE = 50_000_000
array_a = np.random.rand(ARRAY_SIZE)
array_b = np.random.rand(ARRAY_SIZE)

print("\nEtapa 2: Dois arrays NumPy de 50 milhões de elementos criados.")

# ---

# 3. Abordagem Tradicional com Loop (Python Puro)
# Vamos calcular a soma dos quadrados de cada elemento, um por um, usando um loop.
def sum_of_squares_python_loop(a, b):
    result = 0
    for i in range(len(a)):
        result += a[i]**2 + b[i]**2
    return result

# Medir o tempo de execução
start_time = time.time()
sum_py = sum_of_squares_python_loop(array_a, array_b)
end_time = time.time()

time_py = end_time - start_time
print("\nEtapa 3: Cálculo com loop Python tradicional concluído.")
print(f"Resultado do loop Python: {sum_py:.4f}")
print(f"Tempo de execução (Python puro): {time_py:.4f} segundos")
```
