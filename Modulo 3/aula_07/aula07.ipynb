# Aula 07 - Gráficos com Matplotlib

```python
import matplotlib.pyplot as plt

copasBrasil = {
    1958 : "Suécia",
    1962 : "Chile",
    1970 : "México",
    1994 : "Estados Unidos",
    2002 : 'Coreia/Japão',
}

anos = copasBrasil.keys()
pais = copasBrasil.values()



plt.figure(figsize=(6 ,3), facecolor='#012169')
plt.title('Anos que o Brasil foi campeão da copa do mundo', color = "#fedd00")
plt.plot(pais, anos, color="#009739")# gráficos de linha , pais =x, anos = y
plt.xlabel('País sede', color = "#fedd00")
plt.ylabel('Ano', color = "#fedd00")

plt.show()
```

**Grafico de colunas**

```python
meses = ['Janeiro','Fevereiro','Março','Abril','Maio','Junho']
vendas = [150,285,682,450,500,550]

plt.figure(figsize=(6,4))
plt.bar(meses,vendas,color=['black','grey'])
plt.title('Quantidade de vendas - Primeiro semestre de 2026')

plt.show()
```

```python
import pandas as pd

df = pd.read_excel('Base_vendas.xlsx')

df['Faturamento'] = df ['Qtd Vendida'] * df['Preço Unitário']


display(df.head())


for i in df.columns:
    print(i)
```

```python
faturamentoLoja = df.groupby('Loja')['Faturamento'].sum().sort_values()

display(faturamentoLoja)

plt.figure(figsize=(15,4))
faturamentoLoja.plot(kind='barh')

plt.show()
```

```python
fatGenero = df.groupby('Gênero')['Qtd Vendida'].sum().sort_values(ascending=False)


display(fatGenero)

plt.figure(figsize=(15,5),)
fatGenero.plot(kind='pie', colors=['Blue','#c11c84'], autopct='%1.1f%%')
plt.title('Quantidade vendida por gênero')
plt.legend()

plt.show()
```
