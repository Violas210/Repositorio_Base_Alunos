### Você foi contratado como analista de dados e sua primeira atividade:

* Qual o faturamento da empresa ?  (vamos fazer juntos!)
* Quais produtos são vendidos ?
* Qual o produto mais vendido?
* Como foi o faturamento das vendass por gênero ?
* Relacione as vendas por Loja.

```python
import pandas as pd

df = pd.read_excel('Base_vendas.xlsx')

display(df.head())

for i in df.columns:
    print(i)
```

```python
# criar a coluna faturamento
# Qtd Vendida * Preço Unitário

df['Faturamento'] = df['Qtd Vendida'] * df['Preço Unitário']

display(df[['Produto', 'Qtd Vendida', 'Preço Unitário','Faturamento']].head())
```

```python
produtosVndd = df['Produto'].unique()
for i in produtosVndd:
    print(i)
```

```python
produtoMaisVndd = df.groupby(['Produto'])['Qtd Vendida'].sum().sort_values(ascending=False).head(1)
print(produtoMaisVndd)
```

```python
fatGenero = df.groupby('Gênero')['Faturamento'].sum()
print(fatGenero)
```

```python
fatLoja = df.groupby('Loja')['Faturamento'].sum().sort_index()
print(fatLoja)
```
