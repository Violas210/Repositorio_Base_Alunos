```python
# caso precise
# %pip install pandas

import pandas as pd

# criar o DATA FRAME

df = pd.read_csv('tabela-fipe-historico-precos.csv')

display(df.head(10))

print(df.shape)

print(df.info())

print(df.columns)
```

```python
# Qual o carro mais caro de toda a série histórica

# função "loc" localiza não apenas o campo, mas o registro inteiro
maisCaro = df.loc[df['valor'].idxmax()]

print(maisCaro)
```

```python
# nunique - conta registros "ÚNICOS"# unique - apresenta apenas registros únicos
# quantas marcas já operaram no Brasil.

qtdeMarcas = df['marca'].nunique()
print(f'Entre 1992 até 2022 {qtdeMarcas} marcas operaram no Brasil')

qtdeModelo = df['modelo'].nunique()
print(f'Entre 1992 até 2022 {qtdeModelo} modelos operaram no Brasil')
```

```python
# unique - apresenta apenas registros únicos

marcas = df['marca'].unique()

print(marcas)
```

```python
# criar ranking top 10 marcas mais valiosas

marcasTop10 = round(df.groupby('marca')['valor'].mean().sort_values(ascending=False).head(20))

print(marcasTop10)
```

```python
# Quais são os modelos dos 30 carros mais carros da série histórica

top30Modelos = df.groupby('modelo')['valor'].max().sort_values(ascending=False).head(30)

#print(top30Modelos)
print(top30Modelos.map(lambda x: f"R$ {x:,.2f}".replace(',','x').replace('.',',').replace('x','.')))

# explicação próxima aula !
```
