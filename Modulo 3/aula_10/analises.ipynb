```python
import pandas as pd
df = pd.read_csv('Compras.csv', sep=';')

for coluna in df.columns:
    print(coluna)

```

# KPI'S

```python
maisCaro = df.loc[df['ValorFinal'].idxmax()]
print(maisCaro)
```

```python
# Quais produtos foram mais comprados

maisCompr = df.groupby('Produto')['Quantidade'].sum().sort_values(ascending=False).head()

print(maisCompr)
```

```python
vendedorVendMais = df.groupby('Fornecedor')['ValorFinal'].sum().sort_values(ascending=False).head(1)

print(vendedorVendMais)
```

```python
dataMaisVendeu =  df.groupby('Data')['Quantidade'].sum().sort_values(ascending=False).head(1)
print(dataMaisVendeu)
```

```python
top1Fornecedor = df.groupby('Fornecedor')['Quantidade'].sum().idxmax()
print(f'O Fornecedor que mais vendeu foi: {top1Fornecedor}')
```

```python
df['Regiao'] = df['Fornecedor'].str.split().str[-1]

display(df.head(20))
```

```python
faturamnento_total = df['ValorFinal'].sum()

print(f'Faturamento total foi {faturamnento_total}')
```
